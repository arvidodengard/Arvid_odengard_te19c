```python
# Uppgift 2 a,b,c,d,e
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kf = pd.read_csv("sysselsattning.csv", encoding="ISO_8859-1", header=1)
kf_G = kf[kf["sysselsättning"]=="förvärvsarbetande"]
kf_K = kf_G[kf_G["kön"]=="män"]
kf_sk = kf_K.sum(axis = 1).sum()
print(f"Antalet män som förvärvsarbetade är {kf_sk}")

kf_G = kf[kf["sysselsättning"]=="förvärvsarbetande"]
kf_K = kf_G[kf_G["kön"]=="kvinnor"]
kf_sp = kf_K.sum(axis = 1).sum()
print(f"Antalet kvinnor som förvärvsarbetade är {kf_sp}")

kf_G = kf[kf["sysselsättning"]=="ej förvärvsarbetande"]
kf_K = kf_G[kf_G["kön"]=="män"]
kf_sa = kf_K.sum(axis = 1).sum()
print(f"Antalet män som ej är förvärvsarbetade {kf_sa}")

kf_G = kf[kf["sysselsättning"]=="ej förvärvsarbetande"]
kf_K = kf_G[kf_G["kön"]=="kvinnor"]
kf_sl = kf_K.sum(axis = 1).sum()
print(f"Antalet kvinnor som ej är förvärvsarbetade {kf_sl}")


förvärvsarbetare = [kf_sk, kf_sp, kf_sa, kf_sl]

labels = ["m f", "k f", "m ej f", "k ej f"]

plt.bar(labels, förvärvsarbetare, color = ["green", "blue","red","black"])

plt.title("m = män k = kvinnor f = förvärvsarbetande ")
plt.ylabel ("miljoner")
plt.show()
```
